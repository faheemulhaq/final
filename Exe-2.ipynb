```python
titles = pd.read_csv('titles.csv',index_col=None)
titles.head()
```

```python
cast = pd.read_csv('cast.csv', index_col=None)
cast.head()
```

```python
release = pd.read_csv('release_dates.csv',index_col=None)
release.head()
```

```python
title_counts = cast['title'].value_counts()

# Sort the titles by counts in descending order
sorted_titles = title_counts.sort_values(ascending=False)

# Display the ten most common movie titles
top_10_titles = sorted_titles.head(10)
print(top_10_titles)
```

```python
movies_1930s = release[(release['year'] >= 1930) & (release['year'] <= 1939)]

# Count the number of films released per year
films_per_year = movies_1930s['year'].value_counts()

# Sort the counts in descending order
sorted_years = films_per_year.sort_values(ascending=False)

# Identify the top three years with the most films released
top_3_years = sorted_years.head(3)
print(top_3_years)
```

```python
release['decade'] = (release['year'] // 10) * 10

# Group by decade and count the number of films
films_per_decade = release.groupby('decade').size()

# Plotting
plt.figure(figsize=(10, 6))
plt.bar(films_per_decade.index, films_per_decade.values, width=8, color='skyblue')
plt.xlabel('Decade')
plt.ylabel('Number of Films Released')
plt.title('Number of Films Released Each Decade')
plt.xticks(films_per_decade.index)
plt.grid(axis='y')
plt.show()
```

```python
hamlet_movies = titles[titles['title'].str.contains('Hamlet', case=False) & titles['year'].notna()]

# Convert 'year' to numeric (assuming it's an integer column representing the release year)
hamlet_movies['decade'] = (hamlet_movies['year'] // 10) * 10

# Group by decade and count the number of Hamlet movies
hamlet_per_decade = hamlet_movies.groupby('decade').size()

# Plotting
plt.figure(figsize=(10, 6))
plt.bar(hamlet_per_decade.index, hamlet_per_decade.values, width=8, color='skyblue')
plt.xlabel('Decade')
plt.ylabel('Number of "Hamlet" Films Released')
plt.title('Number of "Hamlet" Films Released Each Decade')
plt.xticks(hamlet_per_decade.index)
plt.grid(axis='y')
plt.show()
```

```python
character_counts = cast['name'].value_counts()

# Sort the characters by counts in descending order
sorted_characters = character_counts.sort_values(ascending=False)

# Display the 11 most common character names
top_11_characters = sorted_characters.head(11)
print(top_11_characters)
```

```python
himself_credits = cast[cast['character'].str.lower() == 'himself']

# Count occurrences of each actor credited as "Himself"
actor_counts = himself_credits['name'].value_counts()

# Sort the actors by counts in descending order
sorted_actors = actor_counts.sort_values(ascending=False)

# Display the top 10 actors credited as "Himself"
top_10_himself_actors = sorted_actors.head(10)
print(top_10_himself_actors)
```

```python

movies_1945 = cast[cast['year'] == 1945]

    
# Count appearances for each actor in 1945
actor_appearances = movies_1945['character'].value_counts()

# Count appearances for each actress in 1945
actress_appearances = movies_1945['character'].value_counts()

# Identify the actor or actress with the most appearances in 1945
most_appearances_actor = actor_appearances.idxmax()
most_appearances_actress = actress_appearances.idxmax()

# Display the results
print("Actor with the most appearances in 1945:", most_appearances_actor)
print("Actress with the most appearances in 1945:", most_appearances_actress)
```

```python
mammootty_roles = cast[cast['name'] == 'Mammootty']

# Group by year and count the number of roles each year
roles_per_year = mammootty_roles.groupby('year').size()

# Plotting
plt.figure(figsize=(10, 6))
plt.plot(roles_per_year.index, roles_per_year.values, marker='o', color='skyblue')
plt.xlabel('Year')
plt.ylabel('Number of Roles')
plt.title('Number of Roles Played by Mammootty Each Year')
plt.grid(True)
plt.show()
```

```python
patron_in_roles = cast[cast['character'].str.lower().str.startswith('patron in', na=False)]

# Count occurrences of each role starting with "patron in"
role_counts = patron_in_roles['character'].value_counts()

# Sort the roles by counts in descending order
sorted_roles = role_counts.sort_values(ascending=False)

# Display the top 10 most frequent roles starting with "patron in"
top_10_roles = sorted_roles.head(10)
print(top_10_roles)
```

```python
science_roles = cast[cast['character'].str.contains(r'\bscience', regex=True, na=False)]

# Count occurrences of each role starting with "science"
role_counts = science_roles['character'].value_counts()

# Sort the roles by counts in descending order
sorted_roles = role_counts.sort_values(ascending=False)

# Display the top 10 most frequent roles starting with "science"
top_10_roles = sorted_roles.head(10)
print(top_10_roles)
```

```python
judi_dench_roles = cast[cast['name'] == 'Judi Dench']

# Count occurrences of each unique role
role_counts = judi_dench_roles['name'].value_counts()

# Plotting
plt.figure(figsize=(12, 6))
role_counts[:20].plot(kind='bar', color='skyblue')
plt.xlabel('Role')
plt.ylabel('Number of Appearances')
plt.title('Number of Appearances for Each Unique Role (Judi Dench)')
plt.xticks(rotation=45)
plt.show()
```

```python
cary_grant_roles = cast[cast['name'] == 'Cary Grant']

# Count occurrences of each unique role
role_counts = cary_grant_roles['name'].value_counts()

# Plotting
plt.figure(figsize=(12, 6))
role_counts[:20].plot(kind='bar', color='skyblue')
plt.xlabel('Role')
plt.ylabel('Number of Appearances')
plt.title('Number of Appearances for Each Unique Role (Cary Grant)')
plt.xticks(rotation=45)
plt.show()
```

```python
sidney_poitier_roles = cast[cast['name'] == 'Sidney Poitier']

# Count occurrences of each unique role per year
role_counts_per_year = sidney_poitier_roles.groupby(['year', 'character']).size().unstack()

# Plotting
plt.figure(figsize=(14, 8))
role_counts_per_year.plot(marker='o', linestyle='-', linewidth=2)
plt.xlabel('Year')
plt.ylabel('Number of Appearances')
plt.title('Number of Appearances for Each Unique Role (Sidney Poitier)')
plt.legend(title='Role', bbox_to_anchor=(1, 1), loc='upper left')
plt.grid(True)
plt.show()
```

```python
leading_roles_1950s = cast[(cast['year'] >= 1950) & (cast['year'] < 1960) & (cast['n'] == 1)]

# Count leading roles for actors and actresses
leading_roles_count = leading_roles_1950s['character'].value_counts()

# Separate counts for actors and actresses
actors_count = leading_roles_count[leading_roles_count.index.str.contains('actor', case=False, na=False)].sum()
actresses_count = leading_roles_count[leading_roles_count.index.str.contains('actress', case=False, na=False)].sum()

print("Number of leading roles (n=1) for actors in the 1950s:", actors_count)
print("Number of leading roles (n=1) for actresses in the 1950s:", actresses_count)
```
