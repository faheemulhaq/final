# Write a Python program to find a target values in a list using linear search with following steps: 
a. Initialize the list to store the input elements.

b. Initialize found-False.

c. Enter the item to be searched (match_item),

d. For each element in the list

1. if match item = value a. return match item's position.

e. If the match item is not in the list, display an error message that the item is not found in

the list.

```python
def linear_search(arr, match_item):
    found = False
    for index, value in enumerate(arr):
        if value == match_item:
            return index
            found = True

    if not found:
        print("Item not found in the list.")

# Initialize the list
input_list = []

# Input the elements of the list
for i in range(5):
    element = int(input(f"Enter element {i+1}: "))
    input_list.append(element)

# Input item to be searched
match_item = int(input("Enter the item to be searched: "))

# Perform linear search
result = linear_search(input_list, match_item)

if result is not None:
    print(f"Item found at index {result}.")
```

# Write a Python script to perform the following operations on a singly linked list

a. Create a list

b. Find the smallest element from the list

c. Insert an element if it is not a duplicate element

d. Display the elements in reverse order

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

class LinkedList:
    def __init__(self):
        self.head = None

    def insert(self, data):
        new_node = Node(data)
        if not self.head:
            self.head = new_node
        else:
            temp = self.head
            while temp.next:
                temp = temp.next
            temp.next = new_node

    def display(self):
        if not self.head:
            print("The list is empty.")
            return
        current = self.head
        while current:
            print(current.data, end=" ")
            current = current.next
        print()

    def find_smallest(self):
        if not self.head:
            print("The list is empty.")
            return None
        smallest = self.head.data
        current = self.head.next
        while current:
            if current.data < smallest:
                smallest = current.data
            current = current.next
        return smallest

    def insert_unique(self, data):
        if not self.head:
            self.head = Node(data)
            return

        current = self.head
        while current:
            if current.data == data:
                print(f"{data} is a duplicate, not inserting.")
                return
            if current.next is None:
                current.next = Node(data)
                return
            current = current.next

    def display_reverse(self, node):
        if node:
            self.display_reverse(node.next)
            print(node.data, end=" ")


# Creating a linked list and performing operations
linked_list = LinkedList()

n = int(input("Enter the number of elements: "))

for i in range(n):
    data = int(input(f"Enter element {i+1}: "))
    linked_list.insert(data)
    

print("List elements:")
linked_list.display()

print("Smallest element:", linked_list.find_smallest())

linked_list.insert_unique(40)  # Inserting a unique element
linked_list.insert_unique(10)  # Attempting to insert a duplicate

print("List elements after insertions:")
linked_list.display()

print("List elements in reverse:")
linked_list.display_reverse(linked_list.head)
```

# Write a Python program to implement the various operations for Stack ADT i.) insert ii.) delete iii.) Display.

```python
 class Stack:
    def __init__(self):
        self.stack = []

    def is_empty(self):
        return len(self.stack) == 0

    def push(self, item):
        self.stack.append(item)
        print(f"Pushed {item} onto the stack.")

    def pop(self):
        if not self.is_empty():
            item = self.stack.pop()
            print(f"Popped {item} from the stack.")
            return item
        else:
            print("Stack is empty. Cannot pop.")

    def display(self):
        if not self.is_empty():
            print("Stack elements:")
            for item in reversed(self.stack):
                print(item)
        else:
            print("Stack is empty. Nothing to display.")


# Create a stack and perform operations
stack = Stack()

n = int(input("Enter the number of elements: "))

for i in range(n):
    data = int(input(f"Enter element {i+1}: "))
    stack.push(data)

# Display the stack
stack.display()

# Pop an element
stack.pop()

# Display the updated stack
stack.display()
```

# Write a Python program to convert the following infix expression to its postfix using the push and pop operation of a stack
a.  A/B/C+D'E-FG
b. (B^2-4*A*C)^(1/2) (100)

```python
class Stack:
    def __init__(self):
        self.stack = []

    def push(self, item):
        self.stack.append(item)

    def pop(self):
        if not self.is_empty():
            return self.stack.pop()
        else:
            return None

    def peek(self):
        if not self.is_empty():
            return self.stack[-1]
        else:
            return None

    def is_empty(self):
        return len(self.stack) == 0


def precedence(operator):
    precedence = {'+': 1, '-': 1, '*': 2, '/': 2, '^': 3}
    return precedence.get(operator, 0)


def infix_to_postfix(infix_expression):
    output = []
    stack = Stack()

    for char in infix_expression:
        if char.isalnum():
            output.append(char)
        elif char == '(':
            stack.push(char)
        elif char == ')':
            while stack.peek() != '(' and not stack.is_empty():
                output.append(stack.pop())
            stack.pop()  # Pop the '('
        else:
            while (
                not stack.is_empty() and precedence(char) <= precedence(stack.peek())
            ):
                output.append(stack.pop())
            stack.push(char)

    while not stack.is_empty():
        output.append(stack.pop())

    return ''.join(output)


# Example usage
infix_expression_a = "A/B/C+D'E-FG"
infix_expression_b = "(B^2-4*A*C)^(1/2) (100)"

postfix_a = infix_to_postfix(infix_expression_a)
postfix_b = infix_to_postfix(infix_expression_b)

print("Infix expression A:", infix_expression_a)
print("Postfix expression A:", postfix_a)

print("\nInfix expression B:", infix_expression_b)
print("Postfix expression B:", postfix_b)
```
